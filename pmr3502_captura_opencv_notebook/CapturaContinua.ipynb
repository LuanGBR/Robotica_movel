```python
import cv2
from matplotlib import pyplot as plt
from IPython.display import clear_output
```

A célula abaixo captura e exibe imagens até ser interrompida.
(Provavelmente haverá uma permuta entre os canais "azul" e "vermelho")

```python
vid = cv2.VideoCapture(0, cv2.CAP_V4L2)
vid.set(cv2.CAP_PROP_AUTOFOCUS,0)
vid.set(cv2.CAP_PROP_FOCUS,0)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
try:
    while True:
        ret, frame = vid.read()
        if ret:
            clear_output(wait=True)
            frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            plt.imshow(frame)
            plt.show()
except KeyboardInterrupt:
    pass
finally:
    vid.release()
```
